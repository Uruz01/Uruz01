```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Given array of city names
cities = np.array(['New York', 'Lisbon', 'Beijing', 'Quebec'])

# Create a new array with the lengths of each element
city_lengths = np.array([len(city) for city in cities])

# Display the results
print("Cities:", cities)
print("Lengths of Each City Name:", city_lengths)

# Create a DataFrame for better visualization
df = pd.DataFrame({
    'City': cities,
    'Length': city_lengths
})

print("\nCity Name Lengths:")
print(df)

# Plot the lengths with a bar graph
plt.figure(figsize=(8, 5))
plt.bar(cities, city_lengths, color='skyblue')

plt.title("Length of Each City Name")
plt.xlabel("City")
plt.ylabel("Length")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```
